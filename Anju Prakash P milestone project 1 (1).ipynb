```python
#removing outlier of inches

Q1 = df['Inches'].quantile(0.25)
Q3 = df['Inches'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = df[(df['Inches'] >= lower_bound) & (df['Inches'] <= upper_bound)]

# Create a boxplot after removing outliers
plt.figure(figsize=(5, 3))
sns.boxplot(df_no_outliers['Inches'])
plt.title("Boxplot After Removing Outliers of inches ")
plt.show()
```

# Multivariate Analysis

```python
# Create a pivot table with mean Price
pivot_table = df.pivot_table(values='Price', index='Company', columns='TypeName', aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

# Add title and labels
plt.title('Heatmap of Average Price by Company and TypeName')
plt.xlabel('TypeName')
plt.ylabel('Company')

# Show plot
plt.show()
```

The pivot table is created in this context to restructure the data so that it can be effectively visualized as a heatmap. Heatmaps require a grid-like numeric format, where rows and columns represent categorical data, and cell values are numerical. Here's why the pivot table is necessary:

The heatmap maps rows to one category (e.g., Company), columns to another category (e.g., TypeName), and the color intensity represents the numeric value (e.g., Price).
Without the pivot table, the heatmap wouldn't know how to aggregate and display the relationships between categories and numbers.

By creating a pivot table, the heatmap becomes a powerful tool to summarize and visualize multivariate relationships between Company, TypeName, and Price.



**the price range is between 25000 to 200000, but in the razer company which belongs to gaming laptop the price more than 200000.so the razer gaming laptop is most expensive among the given laptops,and the least laptop are coloured in blue.which are less than 25000.**

```python
# Compute correlation matrix
correlation_matrix = df[['Price', 'Weight', 'Inches']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add title
plt.title('Correlation Heatmap for Numeric Features')
plt.show()
```

Price vs. Weight: A value like 0.15 would indicate a moderate positive correlation, meaning heavier laptops tend to have higher prices.
Price vs. Inches: A value like 0.02 would indicate a weak positive correlation, meaning laptops with larger screen sizes tend to cost more.
Weight vs. Inches: A value like 0.08 would indicate a weak positive correlation, meaning larger screens are generally associated with heavier laptops.
